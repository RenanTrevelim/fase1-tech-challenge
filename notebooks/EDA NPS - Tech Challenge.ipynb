### ● Por que o NPS é importante para um e-commerce?

**Resposta:**  
O NPS mede a satisfação e a lealdade do cliente. Em um e-commerce, isso é essencial, pois o cliente pode mudar de loja com facilidade.

Um NPS alto indica maior chance de recompra e recomendação, enquanto um NPS baixo aponta problemas na experiência do cliente.

---

### ● Quais áreas poderiam se beneficiar desses insights?

- **Produto:** avaliar qualidade e expectativa vs. realidade  
- **Pricing:** entender se o preço está alinhado com o valor percebido

---

### ● Como o NPS impacta a recompra, boca a boca e market share em e-commerce?

**Resposta:**  
O NPS impacta diretamente a recompra, pois clientes satisfeitos (promotores) tendem a comprar novamente, enquanto clientes insatisfeitos (detratores) dificilmente retornam.

No boca a boca, clientes com NPS alto costumam recomendar a marca para outras pessoas, gerando crescimento orgânico. Já clientes com NPS baixo podem compartilhar experiências negativas, prejudicando a imagem da empresa.

Em relação ao market share, um NPS elevado contribui para o aumento da base de clientes e fidelização, ajudando a empresa a ganhar espaço no mercado. Por outro lado, um NPS baixo pode resultar em perda de clientes para concorrentes.

## O que é NPS?

O **Net Promoter Score (NPS)** é uma métrica amplamente utilizada para medir **satisfação e lealdade do cliente**. Ela se baseia na pergunta:

> *"Em uma escala de 0 a 10, o quanto você recomendaria nossa empresa ou serviço para um amigo ou colega?"*

A partir dessa avaliação, os clientes são classificados em três categorias:

- **Detratores (0 a <6):** clientes insatisfeitos ou frustrados, com maior probabilidade de gerar feedback negativo ou abandonar o serviço.
- **Neutros (6 a <8):** clientes relativamente satisfeitos, mas que não demonstram forte lealdade à marca.
- **Promotores (8 a 10):** clientes muito satisfeitos, com maior probabilidade de recomendar a empresa.

# 3.4. Fatores mais críticos relacionados ao NPS

```python
def plot_graficos(dados, x, y, titulo_y=None):
    fig, ax = plt.subplots(1, 3, figsize=(16,4))

    palette = {
        'Detratores': '#2ecc71',
        'Neutros': '#b0b0b0',
        'Promotores': '#2ecc71'
    }

    # Boxplot
    sns.boxplot(data=dados, x=x, y=y, ax=ax[0], palette=palette, hue=x, legend=False)
    ax[0].set_title(f'Boxplot de {y}')
    ax[0].set_xlabel(x)
    ax[0].set_ylabel(y)

    # Violinplot
    sns.violinplot(data=dados, x=x, y=y, ax=ax[1], palette=palette, hue=x, legend=False)
    ax[1].set_title(f'ViolinPlot de {y}')
    ax[1].set_xlabel(x)
    ax[1].set_ylabel(y)

    # KDE plot
    sns.kdeplot(data=dados, x=y, hue=x, ax=ax[2], fill=True)
    ax[2].set_title(f'Densidade de {y}')
    ax[2].set_xlabel(y)

    plt.tight_layout()
    plt.show()
```

Análise da relação entre atraso na entrega e satisfação do cliente.

**Principais insights:**
- Detratores apresentam maior atraso médio
- Promotores concentram-se em entregas sem atraso
- Relação clara entre atraso e queda no NPS

**Impacto no projeto:**
- Atraso na entrega é um dos principais drivers de insatisfação

# 3.6. Feature Engineering

Criação de variáveis derivadas para representar situações críticas na jornada do cliente.

**Variáveis criadas:**
- `entrega_problematica`: identifica casos com **atraso na entrega e múltiplas tentativas**
- `cliente_muito_acionado`: identifica clientes com alto volume de contatos no atendimento

**Impacto no projeto:**
- Foco em cenários mais críticos e específicos
- Redução de ruído ao considerar apenas casos mais severos

```python
dados['entrega_problematica'] = ((dados['atraso_entrega_dias'] > 0) & (dados['tentativas_entrega'] > 1)).astype(int)
dados['cliente_muito_acionado'] = (dados['contatos_atendimento'] >= 3).astype(int)
```

```python
def plot_engineering(dados, x, y, titulo_y=None):
    fig, ax = plt.subplots(1, 3, figsize=(16,4))

    palette = {
        1: '#2ecc71',
        0: '#b0b0b0'
    }

    # Barplot
    sns.barplot(data=dados, x=x, y=y, ax=ax[0],legend=False)
    ax[0].set_title(f'Barplot de {x}')
    ax[0].set_xlabel(x)
    ax[0].set_ylabel(y)

    # Boxplot
    sns.boxplot(data=dados, x=x, y=y, ax=ax[1], palette=palette, hue=x, legend=False)
    ax[1].set_title(f'Boxplot de {x}')
    ax[1].set_xlabel(x)
    ax[1].set_ylabel(y)

    # Violinplot
    sns.violinplot(data=dados, x=x , y=y, hue=x, ax=ax[2], palette=palette,legend=False)
    ax[2].set_title(f'Violinplot de {x}')
    ax[2].set_xlabel(y)

    plt.tight_layout()
    plt.show()
```

### Impacto da Entrega Problemática no NPS

Avaliação do efeito de falhas logísticas críticas sobre a satisfação.

**Principais insights:**
- Casos com entrega problemática apresentam NPS mais baixo
- Diferença existe, mas é **menos acentuada**
- Indica que apenas casos mais graves foram capturados

**Impacto no projeto:**
- Variável mais específica, porém menos abrangente
- Pode reduzir sensibilidade do modelo, mas aumenta precisão em casos críticos

```python
plot_engineering(dados, 'entrega_problematica', 'nps')
```

### Impacto do Cliente Muito Acionado no NPS

# 3.7. Ponto de Ruptura

Análise do impacto progressivo do atraso na entrega sobre o NPS.

**Principais insights:**
- Existe uma relação clara e contínua: quanto maior o atraso, menor o NPS
- Queda acentuada do NPS a partir de **~2 dias de atraso**
- A partir de **4 dias**, o NPS entra em níveis críticos
- Acima de **6 dias**, o NPS se aproxima de zero

**Impacto no projeto:**
- Identificação de um **limite crítico operacional (~2 dias)**
- Atrasos acima desse ponto aumentam drasticamente o risco de detratores
- Permite definição de SLA e priorização de entregas em risco

```python
dados['faixa_atraso'] = pd.cut(
    dados['atraso_entrega_dias'],
    bins=10  # pode ajustar
)
```

```python
analise = (dados.groupby('faixa_atraso', observed=True)['nps'].mean().dropna().reset_index())
```

```python
plt.figure(figsize=(10,5))

plt.plot(analise['faixa_atraso'].astype(str), analise['nps'], marker='o')

plt.xticks(rotation=45)
plt.title('NPS por Faixa de Atraso')
plt.xlabel('Faixa de Atraso (dias)')
plt.ylabel('NPS')

plt.grid()
plt.tight_layout()
plt.show()
```

## Existe algum “ponto de ruptura” na experiência do cliente?

Sim, a análise indica a existência de um ponto de ruptura claro na experiência do cliente, associado ao atraso na entrega.

**Principais evidências:**
- O NPS diminui progressivamente conforme o atraso aumenta
- A queda se torna mais acentuada a partir de **~2 dias de atraso**
- Acima de **4 dias**, o NPS atinge níveis críticos
- Em atrasos superiores a **6 dias**, a satisfação se aproxima de zero

**Interpretação:**
Existe um limite de tolerância operacional. Até cerca de 2 dias, o cliente ainda absorve o atraso; a partir desse ponto, a experiência passa a ser percebida como negativa de forma mais intensa.

**Implicação de negócio:**
Atrasos acima de 2 dias representam alto risco de geração de detratores, sendo um ponto chave para ações preventivas e definição de SLA.

# Que tipo de cliente tende a ter NPS mais alto ou mais baixo?

### Clientes com NPS mais alto (Promotores)
Tendem a apresentar:
- Alto **CSAT interno**
- Entregas sem atraso ou dentro do prazo
- Baixo número de reclamações
- Pouco ou nenhum contato com o atendimento
- Maior probabilidade de recompra

👉 Representam uma experiência fluida, sem fricções operacionais ou de atendimento.

---

### Clientes com NPS mais baixo (Detratores)
Tendem a apresentar:
- Baixo **CSAT interno**
- Atrasos na entrega (especialmente acima de 2 dias)
- Alto número de reclamações
- Múltiplos contatos com o atendimento
- Baixa probabilidade de recompra

👉 Representam uma jornada com falhas acumuladas, tanto operacionais quanto de experiência.
