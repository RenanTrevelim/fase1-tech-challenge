### ● Por que o NPS é importante para um e-commerce?

O NPS é um indicador direto da qualidade da experiência do cliente e da sua propensão a continuar comprando.

O NPS está diretamente ligado ao crescimento do negócio.

- Clientes satisfeitos tendem a recomprar e recomendar a marca

Dessa forma, o NPS influencia diretamente:
- **Recompra:** retenção de clientes

# 2. Definição da Target

Nesta etapa, define-se a variável alvo do problema, alinhando o conceito de satisfação do cliente com a modelagem preditiva.

---

### ● Qual variável representa a satisfação do cliente?

**Resposta:**  
A variável que representa a satisfação do cliente é o **NPS (Net Promoter Score)**.

---

### ● Por que ela foi escolhida?

**Resposta:**  
O NPS foi escolhido por ser um dos principais indicadores de experiência do cliente, amplamente utilizado no mercado.

Ele sintetiza a percepção do cliente sobre toda a jornada de compra e permite classificá-lo em diferentes níveis de satisfação (promotores, neutros e detratores), sendo diretamente relacionado à retenção e ao crescimento do negócio.

---

### ● Em que momento da jornada essa informação é coletada?

**Resposta:**  
O NPS é coletado ao final da jornada, após a entrega do pedido e a conclusão da experiência do cliente.

Isso significa que ele representa o resultado final da interação do cliente com a empresa.

---

### ● Existe algum risco de usar essa variável de forma inadequada?

**Resposta:**  
Sim. O principal risco está relacionado ao uso de informações que não estariam disponíveis no momento da previsão.

Como o NPS é coletado apenas ao final da jornada, ele não pode ser utilizado como variável explicativa no modelo, sob risco de gerar **data leakage**.

Além disso, variáveis fortemente relacionadas ao resultado final da experiência (como avaliações ou indicadores pós-entrega) devem ser analisadas com cautela, garantindo que o modelo utilize apenas informações disponíveis antes da ocorrência da satisfação.

#### Insight:

O NPS deve ser tratado como variável alvo (resultado), enquanto o modelo deve ser construído com base em dados operacionais da jornada, permitindo antecipar a insatisfação de forma confiável.

## O que é NPS?

O **Net Promoter Score (NPS)** é uma métrica amplamente utilizada para medir a satisfação e a lealdade do cliente.

Ele se baseia na pergunta:

> *"Em uma escala de 0 a 10, o quanto você recomendaria nossa empresa ou serviço?"*

A partir das respostas, os clientes são classificados em:

- **Detratores (0 a 6):** clientes insatisfeitos, com maior risco de churn  
- **Neutros (6 a 8):** clientes satisfeitos, mas pouco engajados  
- **Promotores (8 a 10):** clientes satisfeitos e propensos a recomendar  

---

# 3.4. Fatores mais críticos relacionados ao NPS

```python
def plot_graficos(dados, x, y, titulo_y=None):
    fig, ax = plt.subplots(1, 3, figsize=(16,4))

    palette = {
        'Detratores': '#2ecc71',
        'Neutros': '#b0b0b0',
        'Promotores': '#2ecc71'
    }

    # Boxplot
    sns.boxplot(data=dados, x=x, y=y, ax=ax[0], palette=palette, hue=x, legend=False)
    ax[0].set_title(f'Boxplot de {y}')
    ax[0].set_xlabel(x)
    ax[0].set_ylabel(y)

    # Violinplot
    sns.violinplot(data=dados, x=x, y=y, ax=ax[1], palette=palette, hue=x, legend=False)
    ax[1].set_title(f'ViolinPlot de {y}')
    ax[1].set_xlabel(x)
    ax[1].set_ylabel(y)

    # KDE plot
    sns.kdeplot(data=dados, x=y, hue=x, ax=ax[2], fill=True)
    ax[2].set_title(f'Densidade de {y}')
    ax[2].set_xlabel(y)

    plt.tight_layout()
    plt.show()
```

## Impacto do atraso na entrega

Clientes detratores apresentam, em média, maiores atrasos e maior variação no tempo de entrega.

Já os promotores concentram-se em pedidos com pouco ou nenhum atraso, indicando uma experiência mais previsível.

Essa diferença mostra que o atraso não é apenas um detalhe operacional, mas um fator diretamente ligado à insatisfação do cliente.


```python
# Plotando o 'Atraso_entrega_dias'
plot_graficos(dados, 'classe', 'atraso_entrega_dias')
```

### Insight

A inconsistência na entrega — tanto em atrasos quanto na variabilidade — está fortemente associada à geração de detratores.

### Direcionamento

Reduzir atrasos deve ser uma prioridade operacional.

Atuar principalmente nos casos mais críticos e na variabilidade dos prazos tende a gerar impacto direto na redução de detratores.

## Impacto do número de reclamações

Clientes detratores apresentam maior volume de reclamações e também maior variação, indicando recorrência de problemas ao longo da jornada.

Já os promotores concentram-se em níveis mais baixos, com menor dispersão.

Isso mostra que não é apenas a ocorrência de problemas, mas a repetição deles que está ligada à insatisfação.


```python
plot_graficos(dados, 'classe', 'numero_reclamacoes')
```

### Insight

A frequência de falhas na experiência tem impacto direto no NPS.

Quanto maior o número de reclamações — especialmente recorrentes — maior a tendência de o cliente se tornar detrator.

Reduzir a reincidência de problemas deve ser prioridade.

Atuar na causa raiz e na eficiência da resolução tende a diminuir o volume de reclamações e, consequentemente, a geração de detratores.

# 3.6. Ponto de Ruptura

##  Ponto de ruptura na experiência do cliente

Esta análise identifica a partir de qual nível de atraso na entrega a satisfação do cliente começa a cair de forma relevante.


```python
dados['faixa_atraso'] = pd.cut(
    dados['atraso_entrega_dias'],
    bins=[-1, 0, 2, 4, 6, 10],
    labels=['Sem atraso', 'Até 2 dias', '2-4 dias', '4-6 dias', 'Acima de 6 dias']
)
```

```python
analise = (dados.groupby('faixa_atraso', observed=True)['nps'].mean().dropna().reset_index())
```

```python
plt.figure(figsize=(10,5))

plt.plot(analise['faixa_atraso'].astype(str), analise['nps'], marker='o')

plt.xticks(rotation=45)
plt.title('NPS por Faixa de Atraso')
plt.xlabel('Faixa de Atraso (dias)')
plt.ylabel('NPS')

plt.grid()
plt.tight_layout()
plt.show()
```

##  Existe algum ponto de ruptura na experiência do cliente?

Sim. A análise evidencia um ponto claro de ruptura na experiência do cliente relacionado ao atraso na entrega.


###  Principais insights

Observa-se uma queda consistente do NPS conforme o atraso aumenta, indicando uma relação direta e progressiva entre prazo de entrega e satisfação.

O ponto crítico ocorre a partir de **2 dias de atraso**, onde a queda se torna mais acentuada e contínua. A partir desse nível, cada incremento no atraso gera deterioração significativa na percepção do cliente.

Além disso, atrasos superiores a **4 dias** apresentam impacto ainda mais severo, aproximando o NPS de níveis críticos.

---

###  Implicação para o negócio

O prazo de **até 2 dias de atraso** representa o limite operacional aceitável. A partir desse ponto, há risco elevado de conversão de clientes em detratores.

**Priorizar a redução de atrasos acima de 2 dias — especialmente evitando casos acima de 4 dias — tende a gerar o maior impacto na melhoria da satisfação e do NPS.**


# 3.7. Feature Engineering

### Impacto das variáveis derivadas na satisfação

Foram criadas variáveis derivadas com o objetivo de representar eventos críticos na experiência do cliente, especialmente relacionados à logística, atendimento e recorrência de problemas.

A análise abaixo compara o impacto desses fatores no nível de satisfação dos clientes.

```python
dados['atraso_critico'] = (dados['atraso_entrega_dias'] >= 2).astype(int)

dados['problema_complexo'] = (
    (dados['contatos_atendimento'] >= 2) &
    (dados['tempo_resolucao_dias'] >= 3)
).astype(int)

dados['reclamacao_recorrente'] = (dados['numero_reclamacoes'] >= 3).astype(int)
```

```python
plt.figure(figsize=(16,4))

palette = {
        1: '#2ecc71',
        0: '#b0b0b0'
    }

plt.subplot(1,3,1)
sns.boxplot(x='atraso_critico', y='nps', hue='atraso_critico',
            data=dados, palette=palette, order=[0,1], legend=False)
plt.title('Atraso crítico')
plt.xticks([0,1], ['Não', 'Sim'])

plt.subplot(1,3,2)
sns.boxplot(x='problema_complexo', y='nps', hue='problema_complexo',
            data=dados, palette=palette, order=[0,1], legend=False)
plt.title('Problema complexo')
plt.xticks([0,1], ['Não', 'Sim'])

plt.subplot(1,3,3)
sns.boxplot(x='reclamacao_recorrente', y='nps', hue='reclamacao_recorrente',
            data=dados, palette=palette, order=[0,1], legend=False)
plt.title('Reclamação recorrente')
plt.xticks([0,1], ['Não', 'Sim'])

plt.tight_layout()
plt.show()
```

###  Principais insights

As três variáveis analisadas apresentam impacto negativo no NPS, porém com intensidades distintas:

- **Atraso crítico (≥ 2 dias)** já representa uma queda relevante na satisfação, reforçando o papel da logística como fator estrutural da experiência  
- **Problema complexo** (múltiplos contatos + alta demora na resolução) indica falha operacional mais profunda, com impacto adicional no NPS  
- **Reclamação recorrente** apresenta o maior impacto negativo, evidenciando que a repetição de problemas é o principal fator de deterioração da experiência  

---

###  Leitura integrada dos fatores

Os resultados sugerem uma progressão de criticidade:
**falha operacional → atraso → reclamação → recorrência → detrator**

Ou seja, não é apenas o problema em si que gera insatisfação, mas a incapacidade de resolvê-lo de forma definitiva.

---

### Direcionamento

A atuação deve priorizar a identificação antecipada de clientes em risco, especialmente aqueles que já passaram por múltiplas interações ou apresentam sinais de falhas não resolvidas.

Mais do que responder aos problemas, é necessário interromper o ciclo de reincidência, garantindo que cada ocorrência seja tratada de forma definitiva.

Isso implica evoluir a operação para um modelo mais preventivo, com foco em acompanhamento contínuo de casos críticos e maior integração entre áreas responsáveis pela resolução.

# 3.8. Correlação das Variáveis

##  Correlação das variáveis com a satisfação (NPS)

Esta análise identifica quais variáveis operacionais estão mais associadas à variação na satisfação dos clientes, com base na correlação com o NPS.


```python
# Pegando a correlação apenas dos dados numéricos
correlacao = dados.select_dtypes(include=[np.number]).corr().round(2)

# ordenar do maior para o menor
correlacoes = correlacao['nps'].drop('nps').sort_values(ascending=False)

# top 5 positivas
top_positivas = correlacoes.head(5)

# top 5 negativas
top_negativas = correlacoes.tail(5)

# juntar
top_corr = pd.concat([top_positivas, top_negativas])
```

```python
plt.figure(figsize=(10,6))

sns.barplot(
    x=top_corr.values,
    y=top_corr.index,
    hue=top_corr.index,
    palette='coolwarm',
    dodge=False,
    legend=False
)

plt.title('Top 5 Correlações Positivas e Negativas com NPS')
plt.xlabel('Correlação')
plt.ylabel('Variáveis')

plt.axvline(0, color='black', linestyle='--')
plt.tight_layout()
plt.show()
```

###  Principais insights

As variáveis com maior correlação negativa com o NPS estão diretamente ligadas à operação:

- **Atraso na entrega** apresenta a correlação mais forte (~ -0.6), sendo o principal driver de queda na satisfação  
- **Número de reclamações** e **atraso crítico** também possuem forte impacto negativo, reforçando o efeito acumulativo de falhas na jornada  
- **Reclamação recorrente** e **contatos com atendimento** indicam que problemas persistentes e necessidade de suporte intensificam a insatisfação  

Por outro lado, variáveis como **valor do pedido** e **descontos** apresentam correlação positiva, sugerindo que incentivos financeiros podem amenizar, mas não compensar falhas operacionais.

---

###  Leitura estratégica

A intensidade das correlações mostra que fatores operacionais têm impacto significativamente maior na satisfação do que variáveis comerciais.

Ou seja: **resolver problemas gera mais valor do que oferecer benefícios.**

---

###  Implicação para o negócio

A priorização deve estar na redução de falhas operacionais — especialmente atrasos e recorrência de problemas —, pois são os principais determinantes da insatisfação.

**Iniciativas comerciais (ex: descontos) têm efeito limitado se a experiência operacional não for consistente.**

- Entregas dentro do prazo, sem atrasos relevantes  
- Baixo volume de reclamações  
- Pouca ou nenhuma necessidade de contato com atendimento  
- Jornada fluida, com resolução implícita (sem necessidade de intervenção)  

👉 Representam clientes que não foram expostos a falhas operacionais, evidenciando que **consistência na execução é o principal driver de satisfação**.

###  Clientes com NPS mais baixo (Detratores)

Apresentam acúmulo de falhas ao longo da jornada:

- Atrasos na entrega, especialmente acima de **2 dias (ponto crítico)**  
- Alto volume de reclamações  
- Múltiplos contatos com atendimento  
- Problemas recorrentes ou mal resolvidos  

👉 Não é um evento isolado, mas a **repetição e combinação de falhas** que leva à insatisfação.

---

###  Leitura estratégica

Os dados indicam um padrão claro de degradação da experiência:

**Falha operacional → atraso → reclamação → recorrência → detrator**

Ou seja, o cliente não se torna detrator apenas pelo problema inicial, mas pela incapacidade da operação em resolvê-lo de forma definitiva.

---

###  Implicação para o negócio

Melhorar o NPS não depende apenas de evitar falhas pontuais, mas de **eliminar recorrência e garantir resolução eficaz**.

**Prevenir problemas e evitar sua repetição gera mais impacto do que apenas reagir a eles.**
